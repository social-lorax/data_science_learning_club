# Activity 2: Visuals for Exploratory Data Analysis

For Learning Club Activity 2, we will be learning how to create different visuals that help explore a dataset to better understand its contents. There are some "rules" of creating visualizations when they are intended for communicating the result of an analysis with others or for publication, but at this stage of exploratory data analysis, the data visualizations are for us, the analysts. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
demos = pd.read_csv('Demographic_Statistics_By_Zip_Code.csv')
demos.head()
```

```python
zips = pd.read_csv('zips.csv')
```

```python
coded_data = demos.merge(zips,left_on='JURISDICTION NAME',right_on='Zip')
coded_data.head()
```

# $Scatterplot$ $Matrix$

```python
interesting_columns = ['Boro','PERCENT FEMALE','PERCENT WHITE NON HISPANIC','PERCENT PERMANENT RESIDENT ALIEN',
                       'PERCENT RECEIVES PUBLIC ASSISTANCE']
visual = coded_data[interesting_columns].dropna(how='any')
```

```python
sns.set(style="ticks")
sns.pairplot(visual, hue='Boro',height=4)
```

# $Boxplots$

```python
coded_data[['PERCENT MALE','Boro']].boxplot(by='Boro')
```

# $Histograms$

```python
coded_data[['PERCENT US CITIZEN']].hist()
```
