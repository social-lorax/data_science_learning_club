```python
trees = pd.read_csv('2015_Street_Tree_Census.csv')
```

```python
# How many rows of data are in your dataset

print('There are',trees.shape[0],'rows in the dataset')
```

```python
# Preview a sample of rows in your dataset

trees.head()
```

```python
# Refer to a particular row and column to find a value

print('The first value in the last colum is',trees.iloc[0,-1])
```

```python
# For each column, how many rows have data values, and what datatype are they

trees.info()
```

```python
# What constitutes a missing value (null, blank space, NaN, placeholder value, etc.)

print('Missing values are listed as "NaN":')
trees[pd.isna(trees['health'])].head(1)
```

```python
# Descriptive summary statistics for numeric columns 

trees.describe()
```

```python
# Descriptive summary statistics for categorical columns

trees[['created_at','curb_loc','status','health','spc_latin','spc_common','steward','guards','sidewalk','user_type',
       'problems','root_stone','root_grate','root_other','trunk_wire','trnk_light','trnk_other','brch_light',
       'brch_shoe','brch_other','address','zip_city','borough','nta','nta_name','boro_ct','state','x_sp',
       'y_sp','census tract']].describe()
```
